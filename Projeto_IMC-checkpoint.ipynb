Gerando as alturas dos individuos (em metros)

```python
#Gerando dados de média 1.75 e desvio  padrão de  0.30, sendo 2000 pessoas em uma distribuição normalizada (normal)  de probabilidade
import numpy as np
alt = np.random.normal(1.75, 0.30, 2000)
alt
```

Gerendo o peso a partir das alturas

```python
#Escrevendo a função matematica a ser usada para gerar os pesos
from sympy import *
x1 = symbols("alt_i")
x2 = symbols("aleatorios_i")
peso_i = x1*50 + x2
peso_i
```

```python
#criando a lista de pesos a partir das alturas
from random import *
peso = []
for i in alt:
    a = round(float(uniform(-5, 5)),2)
    b = 50*i
    c= a+b 
    peso.append(c)
```

```python
#vendo a Lista de pesos
print(peso[:20])
```

```python
len(peso)
```

Gerando o IMC

```python
#escrevendo a formula matematica de IMC
peso1, altura1 = symbols(['peso','altura'])
IMC_formula = peso1 / (altura1**2)
IMC_formula 
```

```python
IMC=[]
for i,j in zip(peso,alt):
    e= i / (j**2)
    e= round(e, 2)
    IMC.append(e)
```

```python
print(IMC[:20])
```

Gerando os sexos de forma aleatória

Medidas de tendencia central (media)

```python
#calculando amedia de uma coluna na tabela
def media(coluna):
    soma= sum (coluna)
    media = soma / len(coluna)
    media= round(media,2)
    print(media)
```

```python
#calculando a media  de IMC
media(df.IMC)
```

```python
#calculando a media de peso
media(df.peso)
```

```python
#calculando a media da altura
media(df.altura)
```
